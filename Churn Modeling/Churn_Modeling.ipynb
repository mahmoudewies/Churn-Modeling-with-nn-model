# Encodign data

````python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Geography']=le.fit_transform(df['Geography'])
df['Gender']=le.fit_transform(df['Gender'])
````

# Spliting Data

````python
x=df.drop('Exited',axis=1)
y=df['Exited']
````

# StanderScaler

````python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(x)
````

# Make data balanced

````python
from imblearn.over_sampling import SMOTE

smote=SMOTE(sampling_strategy='minority')
x,y=smote.fit_resample(x,y)
y.value_counts()
````

# split to train and test

````python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
````

# LogisticRegression

````python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
````

````python
print(lr.score(x_train,y_train))
print(lr.score(x_test,y_test))
````

````python
y_pred=lr.predict(x_test)
````

````python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
cm
````

````python
from sklearn.metrics import classification_report
cr=classification_report(y_test,y_pred)
print(cr)
````

````python
sns.heatmap(cm,annot=True,fmt='d',cmap='Blues')
````

````python


import matplotlib.pyplot as plt
models = [lr, dt, rf, knn, svc, nb]
model_names = ['Logistic Regression', 'Decision Tree', 'Random Forest', 'KNN', 'SVC', 'Naive Bayes']
train_scores = []
test_scores = []

for model in models:
  train_scores.append(model.score(x_train, y_train))
  test_scores.append(model.score(x_test, y_test))

plt.figure(figsize=(10, 6))
plt.bar(model_names, train_scores, label='Train Accuracy')
plt.bar(model_names, test_scores, label='Test Accuracy')
plt.xlabel('Models')
plt.ylabel('Accuracy')
plt.title('Model Accuracy Comparison')
plt.xticks(rotation=45, ha='right')
plt.legend()
plt.tight_layout()
plt.show()
````

# Neural network model

````python
import tensorflow as tf
from tensorflow import keras
model=keras.Sequential([
    keras.layers.Dense(128,input_shape=(10,),activation='relu'),
    keras.layers.Dense(64,activation='relu'),
    keras.layers.Dense(1,activation='sigmoid')
])
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
model.fit(x_train,y_train,epochs=100,batch_size=32,validation_data=(x_test,y_test))
````

````python
pd.DataFrame(model.history.history).plot(figsize=(12, 8))
plt.grid(True)
plt.gca().set_ylim(0, 1) # set the vertical range to [0-1]
plt.show()

# Evaluate the model
model_evaluate = model.evaluate(x_test, y_test)
print("Loss     : ",model_evaluate[0])
print("accuracy : ",model_evaluate[1])
````

````python
# Confusion Matrix
y_pred = (model.predict(x_test) > 0.5).astype("int32").tolist()

class_names = ["0", "1"]

# Compute classification report
from sklearn.metrics import classification_report, confusion_matrix
print("Classification report : \n",classification_report(y_test, y_pred, target_names = class_names))

# Function to draw confusion matrix
import seaborn as sns
def draw_confusion_matrix(true, preds, normalize=None):
  # Compute confusion matrix
    conf_matx = confusion_matrix(true, preds, normalize = normalize)
    conf_matx = np.round(conf_matx, 2)
  # print(conf_matx)

  # plt.subplots(figsize=(14,8))
    sns.heatmap(conf_matx, annot=True, annot_kws={"size": 12},fmt="g", cbar=False, cmap="viridis")
    plt.show()

print("Confusion matrix : \n")
draw_confusion_matrix(y_test, y_pred)

print("Normalized confusion matrix : \n")
draw_confusion_matrix(y_test, y_pred, "true")
````
